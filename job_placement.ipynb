### Model

```python
%matplotlib inline

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x= new_df.drop('salary',axis=1)
y= new_df['salary']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
algorithm = ['LinearRegression','DecisionTreeRegressor','RandomForestRegressor','GradientBoostingRegressor']
R2=[]
RMSE = []
```

```python
def models(model):
    model.fit(x_train,y_train)
    pre = model.predict(x_test)
    r2 = r2_score(y_test,pre)
    rmse = np.sqrt(mean_squared_error(y_test,pre))
    R2.append(r2)
    RMSE.append(rmse)
    score = model.score(x_test,y_test)
    print(f'The Score of Model is :{score}')
```

```python
model1 = LinearRegression()
model2 = DecisionTreeRegressor(ccp_alpha=0.5,max_depth = 5)
model3 = RandomForestRegressor(n_estimators=1000)
model4 = GradientBoostingRegressor()
```

```python
models(model1)
models(model2)
models(model3)
models(model4)
```

```python
df1 = pd.DataFrame({'Algorithm':algorithm, 'R2_score': R2, 'RMSE':RMSE})
df1
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(df1.Algorithm,df1.R2_score ,label='R2_score',lw=5,color='peru',marker='v',markersize = 15)
plt.legend(fontsize=15)
plt.show()
```

```python
fig = plt.figure(figsize=(20,8))
plt.plot(df1.Algorithm,df1.RMSE ,label='RMSE',lw=5,color='r',marker='o',markersize = 10)
plt.legend(fontsize=15)
plt.show()
```
